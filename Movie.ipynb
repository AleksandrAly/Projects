### Задача - найти фильм с самым большим количеством указанных жанров после предобработки данных

## Фильмы

```python
import pandas as pd
import numpy as np
from IPython.display import display 
```

```python
name = ['movies_id', 'title', 'genres']
df = pd.read_csv("movies.csv", encoding = 'cp1251', sep = '::', engine = 'python', names = name) 
```

```python
display(df.head(3))
df.tail(3)
```

```python
df.genres.iloc[0]
```

```python
all_genres = np.ravel(df.genres)
all_genres
```

```python
all_genres = []
for x in df.genres:
    all_genres.extend(x.split('|'))
uniq_genres = pd.unique(all_genres)
```

```python
matrix_zero = np.zeros((len(df), len(uniq_genres)))
dummies = pd.DataFrame(matrix_zero, columns = uniq_genres)
dummies
```

```python
#dummies.columns.get_indexer()
list(enumerate(df.genres))
```

```python
list(zip(df.index, df.genres))[:5]
```

```python
for i, gen in enumerate(df.genres):
    id_x = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i, id_x] = 1
dummies
```

```python
movies = df.join(dummies.add_prefix('Genre_'))
movies.sum()
```

```python
movies.head()
```

```python
display(movies.iloc[movies.iloc[:, 3:].sum(axis=1).idxmax(), 1]) # фильм с максимумом жанров
display(movies[movies.iloc[:, 3:].sum(axis=1)>=9].title.values) # фильмы, у которых больше 8 жанров
#movies['summary'] = movies['title'] # пустая колонка
#movies['summary'] = movies.iloc[:, 3:-1].sum(axis=1).astype(int) чтобы не считать заново сумму
movies['summary'] = movies.iloc[:, 3: len(dummies.columns)+3].sum(axis=1).astype(int) # другой способ не учитывать случайно сумму
movies[['title', 'genres', 'summary']]
```
